```python
import matplotlib.pyplot as a
import numpy as np

def square(x):
    return x**2

x_coord_1 = int(input("Введите начальную координату x графика: "))
x_coord_2 = int(input("Введите конечную координату x графика: "))
x_step = int(input("Введите шаг для чисел координаты x (минимальный шаг = 1): "))

x_line = []
y_line = []

x = x_line
y = y_line
f = 0
while_count = 0

for i in range(x_coord_1, x_coord_2 + 1, x_step):
    x_line.append(i)

while while_count < len(x_line):
    while_count += 1
    listp = x_line[f] 
    f += 1
    function_result = square(listp)  
    y_line.append(function_result)
    
a.title("График 1")
a.xlabel("Координата Х")
a.ylabel("Координата Y")
a.plot(x, y)
a.show()
```
